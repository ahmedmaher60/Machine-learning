```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Generating random data for clustering
np.random.seed(0)
X = np.random.randn(150, 2)
```

```python
# Fitting the K-Means model to the data with different number of clusters
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Plotting the elbow method
plt.plot(range(1, 11), sse)
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.grid()
plt.show()
```

```python

# Fitting the K-Means model to the data
kmeans = KMeans(n_clusters=3,init='random',max_iter = 300)
kmeans.fit(X)
```

```python
# Getting the cluster labels for each data point
labels = kmeans.labels_
```

```python
# Plotting the data points and color-coding them based on the assigned clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.show()
```
