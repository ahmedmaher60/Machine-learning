## There are several ways to determine if data needs to be rescaled:

1. Visual inspection: By plotting the data, you can check if the features have different scales and ranges. If the ranges of different features are significantly different, then it's likely that the data needs to be rescaled.

2. Statistical tests: You can use statistical tests to check if the data follows a normal distribution. If the data does not follow a normal distribution, it may need to be rescaled.

3. Algorithm performance: You can check if the performance of the machine learning algorithm is affected by the different scales of the features. If the algorithm's performance is poor, it may be a sign that the data needs to be rescaled.

4. Check the mean and variance of each feature, if they are far from 0 and 1 respectively, that means that you need to normalize or standardize them.

5. Check the correlation matrix of your data, if you have high correlation between two or more features, that means that you need to normalize or standardize them.
