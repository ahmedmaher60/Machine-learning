```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True, normalize=True,copy_X=True,n_jobs=-1)
```

```python
LinearRegressionModel.fit(X_train, y_train)
```

```python
#Calculating Details
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(X_train, y_train)) 
```

```python
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(X_test, y_test))

```

```python
print('Linear Regression Coef is : ' , LinearRegressionModel.coef_) 
```

```python
print('Linear Regression intercept is : ' , LinearRegressionModel.intercept_)
```

```python
# Predicting the Test set results
y_pred = LinearRegressionModel.predict(X_test)
y_pred 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
```

```python
#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # average error of values
print('Mean Absolute Error Value is : ', MAEValue)
```

```python
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)
```

```python
#Calculating Root mean Squared Error
print('Root Mean Squared Error is : ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
```

```python
#Calculating R-squared Error
RValue = r2_score(y_test, y_pred)
print('R-squared Error Value is : ', RValue )
#40 % to 80 % is good value 
```

# Ridge Regression

### Ridge regression is a type of linear regression that adds a regularization term to the cost function. The regularization term is used to shrink the coefficients of the model towards zero. The term is defined as the sum of the squares of the coefficients multiplied by a regularization parameter lambda. The regularization parameter lambda controls the strength of the regularization term.
#### The purpose of the regularization term is to prevent overfitting, which occurs when a model is too complex and is able to fit the noise in the data. Overfitting can lead to poor predictions on new, unseen data. By adding a regularization term to the cost function, ridge regression can help to reduce the variance of the model and improve its generalization performance.

#### One of the key differences between Ridge Regression and normal linear regression is that while in normal linear regression all the features have equal importance, Ridge Regression gives importance to certain features and shrinks the coefficients of others.

#### Ridge Regression is also known as L2 regularization as it adds "squared magnitude" of coefficient as a penalty term to the loss function.
