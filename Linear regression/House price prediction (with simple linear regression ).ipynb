```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\AHMED\\Documents\\Machine Learning\\SKlearn_files-master\\Data\\2.1 Linear Regression\\houses.csv')
```

```python
df = df.dropna()
```

```python
X = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1:]
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y ,test_size=0.30 ,  random_state=44 , shuffle =True )
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error 
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score
```

```python
LinearRegressionModel = LinearRegression(fit_intercept=True, normalize=True,copy_X=True,n_jobs=-1)
```

```python
LinearRegressionModel.fit(X_train, y_train) #help to define theta of equations 
```

## The score() method in scikit-learn's LinearRegression model returns the coefficient of determination (R^2) of the prediction.

> R-squared value, is a measure of how well the model fits the data. It ranges from 0 to 1, where 1 represents a perfect fit and 0 represents a model that does not fit the data at all. A high R-squared value indicates that the model is able to explain a large proportion of the variance in the target variable, while a low R-squared value means that the model is not able to explain much of the variance.

```python
print('Linear Regression Train Score is : ' , LinearRegressionModel.score(X_train, y_train)) #evaluate the performance of a linear regression model.
```

```python
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(X_test, y_test))
```

```python
print('Linear Regression Coef is : ' , LinearRegressionModel.coef_) #cofficient of equations
```

```python
print('Linear Regression intercept is : ' , LinearRegressionModel.intercept_) #intercept with y axis
```

```python
y_pred = LinearRegressionModel.predict(X_test)
print('Predicted Value for Linear Regression is : ' , y_pred[:10])
```

```python
y_test[:10]
```

# Model evaluation

```python
#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # average error of values
print('Mean Absolute Error Value is : ', MAEValue)
```

```python
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
print('Mean Squared Error Value is : ', MSEValue)
```

```python
import numpy as np
#Calculating Root mean Squared Error
print('Root Mean Squared Error is : ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y_test, y_pred)
print('Median Squared Error Value is : ', MdSEValue )
```

```python
#Calculating R-squared Error
RValue = r2_score(y_test, y_pred)
print('R-squared Error Value is : ', RValue )
#40 % to 80 % is good value 
```
